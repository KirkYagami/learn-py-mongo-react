1. **Equality Operator (`'=='`):**

   ```python
   x = 5
   ```

   - Checks if two values are not equal.

   - Checks if the value on the left is less than the value on the right.
   ```python
   x = 5

   result = x < y  # True
   ```

   - Checks if the value on the left is greater than the value on the right.

   - Checks if the value on the left is less than or equal to the value on the right.

   - Checks if the value on the left is greater than or equal to the value on the right.

   ```python
   x = 20
   ```

   - `in` checks if a value exists in a sequence.

The `is` Operator and Integers

The `is` operator checks for **object identity** (whether two variables point to the exact same object in memory), not value equality. For comparing the values of integers, you should always use the equality operator (`==`).

Python's interpreter (specifically the common CPython implementation) pre-allocates and reuses integer objects for a small, specific range, typically from **-5 to 256**, as an optimization.

- For integers _within_ this range, `is` will return `True` for two variables with the same value because they reference the same object.
- For integers _outside_ this range, `is` will typically return `False` because separate objects are created in memory, even if they have the same value.

````python
a = 100000
b = 100000

a is b  # True
# a not b  # False

print("id of a", id(a))
print("id of b", id(b))
````

#### Let's understand `range()`

In Python, the `range()` function generates a sequence of numbers, often used in loops for iteration. By default, it creates numbers starting from 0 up to but not including a specified stop value. You can also reverse the sequence with `reversed()`. If you need to count backwards, then you can use a negative step, like `range(start, stop, -1)`, which counts down from start to stop.

We need to do list(range(5)) to view all the numbers because range() returns a lazy iterable, not a list.

​It generates numbers on demand, and converting it to a list forces it to produce all values immediately.

---
**Note:** Technically, range() is not a function. Instead, range is a type or class. When you call range(), you’re calling the constructor of the range class to create a new range object.

Still, for all practical purposes, you can treat range() as a function that returns a range object.

### **What Makes Something a Sequence?**
In Python, a **sequence** is an ordered collection that supports:

1. **Indexing** - Access elements by position: `seq[i]`
2. **Slicing** - Extract subsequences: `seq[start:stop:step]`
3. **Iteration** - Loop through elements: `for item in seq`
4. **Length** - Count elements: `len(seq)`
5. **Membership testing** - Check containment: `item in seq`

````python
text = "Python"

# 1. Indexing
print("1. Indexing:")
print(f"   text[0] = {text[0]}")
print(f"   text[-1] = {text[-1]}")

# 2. Slicing
print("\n2. Slicing:")
print(f"   text[0:3] = {text[0:3]}")
print(f"   text[::2] = {text[::2]}")

# 3. Iteration
print("\n3. Iteration:")
for i, char in enumerate(text):
    print(f"   Position {i}: {char}")

# 4. Length
print(f"\n4. Length: {len(text)}")

# 5. Membership
print(f"\n5. Membership: 'th' in text = {'th' in text}")
````

Subset

A set `A` is a **subset** of another set `B` if all elements of set `A` are also present in set `B`.
- **Example**: If `A = {1, 2}` and `B = {1, 2, 3, 4}`, `A` is a subset of `B`.
- **Note**: A set is a subset of itself, and the empty set is a subset of every set.

Python Implementation

You can check for a subset relationship using:

- **The `.issubset()` method**: `A.issubset(B)` returns `True` if `A` is a subset of `B`.
- **The `<=` operator**: `A <= B` returns `True` if `A` is a subset of, or equal to, `B`.
- **The `<` operator**: `A < B` checks for a _proper_ subset (all elements of `A` are in `B`, but `A` is not equal to `B`).

```python
A = {1, 2}
B = {1, 2, 3, 4}

print(A.issubset(B))  # Output: True
print(A <= B)         # Output: True
print(A < B)          # Output: True (proper subset)

C = {1, 2, 3, 4}
print(C < B)          # Output: False (not a proper subset, as they are equal)

```


Superset

A set `A` is a **superset** of another set `B` if `A` contains all elements of set `B` (and possibly more). This is the inverse relationship of a subset.

- **Example**: Using the previous sets, `B = {1, 2, 3, 4}` is a superset of `A = {1, 2}`.
- **Note**: A set is a superset of itself.


Python Implementation

You can check for a superset relationship using:

- **The `.issuperset()` method**: `A.issuperset(B)` returns `True` if `A` is a superset of `B`.
- **The `>=` operator**: `A >= B` returns `True` if `A` is a superset of, or equal to, `B`.
- **The `>` operator**: `A > B` checks for a _proper_ superset (all elements of `B` are in `A`, but `A` is not equal to `B`).

```python
A = {1, 2}
B = {1, 2, 3, 4}

print(B.issuperset(A)) # Output: True
print(B >= A)          # Output: True
print(B > A)           # Output: True (proper superset)

C = {1, 2, 3, 4}
print(C > B)           # Output: False (not a proper superset, as they are equal)

```

## What are Comprehensions?

Comprehensions are a Pythonic way to create collections (lists, dictionaries, sets) in a single, readable line. They combine loops and conditional logic into a concise syntax that's both efficient and expressive.

**Basic Syntax Pattern:**

```
[expression for item in iterable if condition]
```


## 1. List Comprehensions

### Definition

Creates a new list by applying an expression to each item in an iterable, optionally filtering with conditions.

### Syntax

```python
[expression for item in iterable if condition]
```

### Example: Square Numbers

````python
# Create a list of squares for numbers 1-5
squares = []
for i in range(1, 6):
    squares.append(i ** 2)
print(squares)  # [1, 4, 9, 16, 25]


````

````python
squares = [i ** 2 for i in range(1, 6)]
print(squares)  # [1, 4, 9, 16, 25]
````

### Use Case Scenarios

- **Data transformation**: Converting strings to uppercase, applying mathematical operations
- **Filtering data**: Getting even numbers, valid email addresses
- **API response processing**: Extracting specific fields from JSON responses
- **File processing**: Reading and transforming file contents


## 2. Dictionary Comprehensions

### Definition

Creates dictionaries using a similar syntax to list comprehensions, allowing you to transform keys and values simultaneously.

### Syntax

```python
{key_expression: value_expression for item in iterable if condition}
```


````python
# TASK: create a hashmap/dict where the key is the word and value is the length

words = ['apple', 'banana', 'cherry', 'date', 'apple']

word_lengths = {}

for word in words:
    word_lengths[word] = len(word)

print(word_lengths)  # {'apple': 5, 'banana': 6, 'cherry': 6, 'date': 4}
````

````python
word_lengths = {word:len(word) for word in words}
print(word_lengths)
````

### Use Case Scenarios

- **Data mapping**: Creating lookup tables, index mappings
- **Configuration parsing**: Converting lists to key-value pairs
- **Database results**: Transforming query results into dictionaries
- **Caching**: Creating quick lookup dictionaries for performance

## 3. Set Comprehensions

### Definition

Creates sets (collections of unique elements) using comprehension syntax, automatically handling duplicates.

### Syntax

```python
{expression for item in iterable if condition}
```

### Example: Unique Word Lengths

````python
words = ['python', 'java', 'go', 'rust', 'php']

unique_lengths = {len(word) for word in words}

print(unique_lengths)  # {5, 4, 3, 4,
````

### Use Case Scenarios

- **Duplicate removal**: Getting unique values from datasets
- **Data validation**: Finding unique identifiers, email domains
- **Set operations**: Preparing data for intersection, union operations
- **Performance optimization**: Fast membership testing

# Lambda Functions in Python

## 1. Introduction to Lambda Functions

Lambda functions are small, anonymous functions that can have any number of arguments but can only have one expression. They're also known as "anonymous functions" because they don't require a `def` statement with a function name.

````python
'''
- any number of args
- ONLY one expression
- no def -> anonymous
- small -> Don't write complex logic inside the lambda

# syntax

lambda arguments: expression

````

````python
f = lambda a,b: a+b
f(2,4)
````

### Components:

- `lambda`: The keyword that defines a lambda function
- `arguments`: Parameters (can be zero or more)
- `:`: Separator between parameters and expression
- `expression`: Single expression that gets evaluated and returned

````python
f2 = lambda : 'Hi, Nick!'
f2()
````

````python
f = lambda a,b,c: a+b+c
f(2,4,4)
````

````python
f3 = lambda *args: sum(args)
f3(1,2,3,4,5,67)
````


### Comparison Table:

| Feature       | Lambda Function        | Regular Function            |
| ------------- | ---------------------- | --------------------------- |
| Definition    | `lambda x: x * 2`      | `def func(x): return x * 2` |
| Name          | Anonymous              | Named                       |
| Statements    | Single expression only | Multiple statements         |
| Documentation | No docstrings          | Can have docstrings         |
| Debugging     | Harder to debug        | Easier to debug             |
| Reusability   | Limited                | High                        |


````python
def calculate_grade(score):
    """Calculate letter grade based on numeric score."""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Lambda - good for simple operations
grade_simple = lambda score: 'Pass' if score >= 60 else 'Fail'

print(calculate_grade(85))  # Output: B
print(grade_simple(85))     # Output: Pass
````

## 1. Introduction to Exception Handling

### What are Exceptions?

**Exceptions** are events that occur during program execution that disrupt the normal flow of instructions. When Python encounters an error, it creates an exception object. If not handled properly, the program terminates abruptly.

````python
num = int(input('Enter the numerator: '))
den = int(input('Enter the denominator: '))

try:
    res = num/den
    print(res)
except ZeroDivisionError:
    print('Denominator cannot be zero')

````

Common built-in exceptions include:

- `ZeroDivisionError`: Occurs when dividing by zero
- `TypeError`: Occurs when an operation is performed on an inappropriate data type
- `ValueError`: Occurs when a function receives an argument of the correct type but inappropriate value
- `NameError`: Occurs when a local or global name is not found
- `IndexError`: Occurs when trying to access an index that is out of range
- `KeyError`: Occurs when a dictionary key is not found
- `FileNotFoundError`: Occurs when trying to open a file that doesn't exist
- `ImportError`: Occurs when an import statement fails
- `AttributeError`: Occurs when an attribute reference or assignment fails


### Why Exception Handling?

Exception handling allows you to:

- **Gracefully manage errors** without crashing your program
- **Separate error-handling code** from regular code
- **Provide meaningful feedback** to users
- **Clean up resources** (files, network connections) even when errors occur
- **Make code more robust and maintainable**

````python
# try except else finally

try:
    num = int(input('Enter the numerator: '))
    den = int(input('Enter the denominator: '))
    res = num/den
except ValueError:
    print('Numerator and denominator should be integers')
except ZeroDivisionError:
    print('Denominator cannot be zero')
except Exception as e:
    print('Something went wrong')
    print(e)
except:
    print('something went wrong')


print('ok...I am safe')
````

### Multiple except Blocks

You can handle different exceptions differently by using multiple `except` blocks.

**Example 3: Multiple Exception Types**

````python
def process_data(data_list, index):
    try:
        # Multiple operations that could fail
        value = data_list[index]
        result = 100 / value
        return result
    except  IndexError:

        print("Errr: Index is out of range")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
    except TypeError:
        print("Error: Invalid data type")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")




````

### Catching Multiple Exceptions Together

When you want to handle multiple exceptions the same way, group them in a tuple.

**Example 4: Grouping Exceptions**

````python
def safe_calculation(a, b, c):
    try:
        result = (a + b) / c
        return result
    except (TypeError, ValueError, ZeroDivisionError) as e:
        print(f"Calculation error occurred: {type(e).__name__}")
        print(f"Error message: {e}")
        return None
````

### Bare except Clause (Not Recommended)

A bare `except` catches ALL exceptions, including system exits and keyboard interrupts.

````python
# NOT RECOMMENDED - Too broad
def risky_bare_except():
    try:
        # Some operation
        value = 10 / 0
    except:
        print("Something went wrong!")

# BETTER - Catch Exception base class
def better_broad_except():
    try:
        value = 10 / 0
    except Exception as e:
        print(f"An error occurred: {e}")
        # This won't catch KeyboardInterrupt or SystemExit

# Testing
better_broad_except()
````

### The else Clause

The `else` block executes **only if no exception was raised** in the try block.

**Example 7: Using else Clause**

````python
# try except else finally
````

````python
def divide_with_else(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Cannot divide by zero!")
        return None
    else:
        # This runs ONLY if no exception occurred
        print("Division successful!")
        return result

# Testing
print(divide_with_else(10, 2))  # Prints "Division successful!" then 5.0
print(divide_with_else(10, 0))  # Prints "Cannot divide by zero!" then None
````

### The finally Clause

The `finally` block **always executes**, regardless of whether an exception occurred or not. It's perfect for cleanup operations.

**Example 9: finally for Cleanup**

````python
def demonstrate_finally():
    try:
        print("1. Trying to open file...")
        file = open('/content/data.txt', 'r')
        print("2. File opened successfully")
        content = file.read()

        # Simulate an error
        result = 10 / 0
    except FileNotFoundError:
        print("3. File not found!")
    except ZeroDivisionError:
        print("3. Division by zero error!")
    else:
        print("4. No errors occurred")
    finally:
        print("5. Finally block - This ALWAYS executes!")
        # Cleanup code here
        try:
            file.close()
            print("6. File closed")
        except:
            print("6. No file to close")

# Testing
demonstrate_finally()
````
