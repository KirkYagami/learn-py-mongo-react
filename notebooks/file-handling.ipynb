## File Open Modes

When opening a file in Python, you specify a mode that determines how you can interact with the file:

|Mode|Description|
|---|---|
|'r'|Read (default) - Open for reading|
|'w'|Write - Create a new file or truncate existing file|
|'a'|Append - Open for writing, appending to the end of file|
|'x'|Exclusive creation - Fail if file already exists|
|'b'|Binary mode (added to other modes)|
|'t'|Text mode (default, added to other modes)|
|'+'|Update mode - Open for reading and writing|


- reading r -> if the file doesnot exist -> Error FileNotFoundError
- writing w -> if the file doesnot exist -> create a new one
- writing w -> if the file exists -> override

## JSON
JSON (JavaScript Object Notation) is a lightweight, text-based, and language-independent data-interchange format. It is used to store and transport data, commonly for transmitting data between a server and a web application. It is human-readable, easy to parse for machines, and structures data in key-value pairs and arrays.

### A. Data Types
1. String: "Hello, World!"
2. Number: 42 or 3.14
3. Boolean: true or false
4. Null: null
5. Object: `{}`
6. Array: `[]`

### B. Objects

- Enclosed in curly braces {}
- Consist of key-value pairs
- Keys must be strings and unique within an object
- Values can be any JSON data type

```python
{
  "name": "John Doe",
  "age": 30,
  "isStudent": false
}
```

### C. Arrays

- Enclosed in square brackets []
- Ordered list of values
- Can contain any JSON data type, including mixed types

```python
["apple", "banana", "cherry", 42, true]
```

## III. Nested Structures

JSON supports complex nested structures:

1. Objects within objects
2. Arrays within objects
3. Objects within arrays
4. Arrays within arrays

```python
{

 "person":{
     "name": "Ru",
     "age": 25,
     "address":{
         "street": "221B",
         "City": "London"
 },
     "hobbies": ["running", "swimming", "singing"],
     "isEmployed": true,
     "Contacts": [
         {"type":"email", "value":"Ru@apple.com"},
         {"type":"phone", "value":"XXXXXXX"}
     ]
 }

}
```

## Python JSON module

- built-in std lib for working with JSON
- it provides functions for converting Python Objects into JSON format <-- serialization
- and converting JSON data back into Python Objects <-- deserialization

![json-fns.png](../assets/images/json-fns.png)

```python
import json
```

```python
python_dict = {
    "name": "Alice",
    "age": 25,
    "city": "New York",
    "is_student": False
}

print(python_dict)
print(type(python_dict))
```

```python
# Response Body

# GET /users/1
# Python Object -> JSON and then transmitting 

json_string = json.dumps(python_dict, indent=4)
print(json_string)
print(type(json_string))
```

```python
# json loads -> des
# JSON strings into python objects
# POST req --> client sends JSON to the server and we need to convert into python objects

back_to_python = json.loads(json_string)

print(back_to_python.get('name'))
print(type(back_to_python))

```
